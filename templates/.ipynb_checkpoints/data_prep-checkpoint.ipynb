```python
import pandas as pd

file_path = r"C:\Users\lebda\Downloads\archive\matches.csv"
df = pd.read_csv(file_path)

print(df.head())
```

```python
df.columns = df.columns.str.strip().str.lower()
print(df.columns)
```

```python
df_h2h = df[["team", "opponent", "gf", "ga", "result"]]
print(df_h2h.head())
```

```python
def match_result(row):
    if row["result"] == 'W':
        return 1
    elif row["result"] == 'L':
        return -1
    else:
        return 0
df["match_result"] = df.apply(match_result, axis=1)
df["match_result"].value_counts()
print(df.head())
```

```python
df_h2h = df[["team", "opponent", "gf", "ga", "match_result"]]
print(df_h2h.head())
```

```python
# Ensure that the dataset is properly sorted before applying rolling calculations
df_h2h = df_h2h.sort_values(by=["team", "opponent"]).reset_index(drop=True)

# Compute rolling averages (needed lots of help with this one to staying within the frame index because it didn't at first)
df_h2h["head_to_head"] = (df_h2h.groupby(["team", "opponent"])["match_result"].transform(lambda x: x.rolling(window=5, min_periods=1).mean()))
df_h2h["head_to_head_goals"] = (df_h2h.groupby(["team", "opponent"])["gf"].transform(lambda x: x.rolling(window=5, min_periods=1).mean()))

df_h2h.head(20)
```

```python
df_h2h.to_csv("head_to_head_updated.csv", index=False)
print("Head-to-head dataset saved")
```
